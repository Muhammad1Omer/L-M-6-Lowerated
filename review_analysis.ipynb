### Rate

```python
from lowerated.rate.entity import Entity
from dotenv import load_dotenv
import os

load_dotenv()

openai_key = os.environ["OPENAI_API_KEY"]

# csv file path
file_path = "dump/Rating System/Sentiment Analysis/Data/movie_review.csv"
#review column
review_column = "text"

# Example usage
if __name__ == "__main__":

    entity = Entity(name="Movie")
    attributes = entity.get_attributes()

    rating = entity.rate(
        file_path=file_path, review_column=review_column, range=(0,100), openai_key=openai_key)

    print(rating)
```
